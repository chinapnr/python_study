```python
from tkinter import *           
root = Tk()                     
                                
list1     = ['java','python','php','js']
list2  = ['mobile','desktop','web']

#  创建两个列表组件
listProgram  = Listbox(root)          
listPlatform = Listbox(root)

# 第一个小部件插入数据
for item in list1:                 
    listProgram.insert(0,item)

# 第二个小部件插入数据
for item in list2:              
    listPlatform.insert(0,item)

# 将小部件放置到主窗口中
listProgram.pack()                    
listPlatform.pack()

# 进入消息循环
root.mainloop()                 
```
